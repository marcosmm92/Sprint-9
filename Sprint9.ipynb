1. Calculadora de l'índex de massa corporal

Escriu una funció que calculi l'IMC ingressat per l'usuari/ària, és a dir, qui ho executi haurà d'ingressar aquestes dades.

La funció ha de classificar el resultat en les seves respectives categories.

Consell: Intenta validar les dades prèviament, perquè enviï un missatge d’advertència si les dades introduïdes per l’usuari/a no estan en el format adequat o no pren valors raonables.

```python
def calculadora_imc():
    """
    Calcula el IMC y lo clasifica según la OMS.
    Datos: peso (kg), altura (m).
    """

    # Validamos el peso
    while True:
        dato_peso = (input("Por favor, introduce tu peso en kg (ej: 75.5): ").replace(",","."))
        try:
            peso = float(dato_peso)
            if peso <= 0:
                print("El peso debe ser un número positivo.")
            elif peso < 20 or peso > 300:
                print("El peso se encuentra fuera del rango habitual. Por favor, compruébalo.")
            else:
                break
        except ValueError:
            print("Entrada errónea. Vuelve a añadirlo correctamente, por favor.")

    # Validamos la altura
    while True:
        dato_altura = (input("Por favor, introduce tu altura en metros (ej: 1.78): ").replace(",","."))
        try:
            altura = float(dato_altura)
            if altura <= 0:
                print("La altura debe ser un número positivo.")
            elif altura < 0.5 or altura > 2.5:
                print("La altura se encuentra fuera del rango habitual. Por favor, compruébala.")
            else:
                break
        except ValueError:
            print("Entrada errónea. Vuelve a añadirla correctamente, por favor.")


    # Calculamos el IMC
    imc = peso / (altura ** 2)

    # Clasificamos según la OMS
    if imc < 18.5:
        categoria = "Bajo peso"
    elif imc < 25:
        categoria = "Peso normal"  
    elif imc < 30:
        categoria = "Sobrepeso"
    else:
        categoria = "Obesidad"

    # Mostramos los resultados
    print(f"Peso: {peso} kg")
    print(f"Altura: {altura} m")
    print(f"\nTu IMC es: {imc:.2f}")
    print(f"Clasificación: {categoria}")

    return imc, categoria

# Ejecutamos la función
calculadora_imc()
```

Existeixen diverses unitats de temperatura utilitzades en diferents contextos i regions. Les més comunes són Celsius (°C), Fahrenheit (°F) i Kelvin (K). També existeixen altres unitats com Rankine (°Ra) i Réaumur (°Re).

Selecciona almenys 2 conversors, de tal manera que en introduir una temperatura retorni, com a mínim, dues conversions, de manera que es puguin guardar (recorda que un print() no es pot guardar mai).

```python
def convertir_temperatura1():
    """
    Convierte las unidades de la temperatura (°C, °F, K y RA).
    Datos: unidad temperatura de origen (string) y valor de la temperatura (float).
    """
    
    # Solicitamos una unidad de temperatura y la validamos
    while True:
        unidad = input("Introduce la unidad de la temperatura (C, F, K, RA): ").strip().upper()
        if unidad in ("C", "F", "K", "RA"):
            break
        print("ERROR: Unidad no válida. Por favor, usa C, F, K o RA.")

    # Solicitamos el valor numérico de la temperatura y lo validamos
    while True:
        try:
            valor = float(input(f"Introduce la temperatura en {unidad}: "))
        except ValueError:
            print("ERROR: Debes introducir un número válido.")
            continue
        break 

    # Conversión de la temperatura   
    match unidad:
        case "C":
            c = valor
            f = c * 9/5 + 32
            k = c + 273.15
            ra = k * 9/5

        case "F":
            f = valor
            c = (f - 32) * 5/9
            k = c + 273.15
            ra = f + 459.67

        case "K":
            k = valor
            c = k - 273.15
            f = c * 9/5 + 32
            ra = k * 9/5

        case "RA":
            ra = valor
            k = ra * 5/9
            c = k - 273.15
            f = ra - 459.67

   # Creamos un diccionario con los resultados
    conversiones = {"Temperaturas": {"Celsius": round(c, 2),"Fahrenheit": round(f, 2),
                                     "Kelvin": round(k, 2),"Rankine": round(ra, 2)}}

    return conversiones

# Ejecutamos la función y mostramos el resultado
resultado = convertir_temperatura1()
print(resultado)            
```

3. Comptador de paraules d'un text.

Escriu una funció que donat un text, mostri les vegades que apareix cada paraula. Intenta que es gestionin totes les casuístiques possibles que facin que el programa no funcioni correctament.

(EXTRA): Quina és la llargada mitjana de les paraules del text que has escrit? “Hola com va?” hauria de retornar (4+3+2) / 3 = 3

```python
def analizar_frase(frase):
    """
    Cuenta las palabras y las letras de un texto.
    Datos: frase, palabras y letras
    """
    
    # Definimos los signos de puntuación que deseamos eliminar
    puntuacion = "?!,.:;¡¿()[]{}\"'"

    # Lo convertimos todo a minúsculas
    frase_limpia = frase.lower()

    # Eliminamos la puntuación
    for signo in puntuacion:
        frase_limpia = frase_limpia.replace(signo, "") 

    # Separamos la frase en palabras
    palabras = frase_limpia.split() 

    # Creamos la estructura para contar
    contador_palabras = {}
    contador_letras = {}
    total_letras = 0

    for palabra in palabras:
        # Contamos cuántas veces aparece cada palabra
        if palabra in contador_palabras:
            contador_palabras[palabra] += 1 
        else:
            contador_palabras[palabra] = 1
        # Contamos las letras de cada palabra
        num_letras = len(palabra) 
        contador_letras[palabra] = num_letras

        # Acumulamos el total de letras
        total_letras += num_letras

    # Calculamos el promedio, evitando que nos dé error si la frase está vacía
    promedio = total_letras / len(palabras) if palabras else 0

    # Devolvemos los resultados
    return contador_palabras, contador_letras, promedio
```

```python
frase = input("Introduce una frase: ")   # Esto mejor añadirlo dentro de la propia función?

conteo, letras, promedio = analizar_frase(frase)

print(f"Conteo de palabras: {conteo}")
print(f"Letras por palabra: {letras}")
print(f"Promedio de letras por palabra: {promedio:.2f}")
```

4. Diccionari invers (amb possibilitat de duplicats)

Resulta que el client té una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al revés, és a dir, intercanviant les claus i els valors. Els valors i claus en el diccionari original són únics; si aquest no és el cas, la funció hauria d'imprimir un missatge d'advertiment, juntament amb una llista amb els valors associats a la clau repetida.

```python
def diccionario_inverso(dic):
    """
    Invierte un diccionario intercambiando claves por valores.

    Si hay valores duplicados, los agrupa en una lista y muestra
    un mensaje de advertencia indicando las claves asociadas.

    Datos: dic (Diccionario original)
    """

    # Creamos los diccionarios donde se guardará el resultado y los valores repetidos
    inverso = {}
    duplicados = {}

    for clave, valor in dic.items():
        # Caso duplicado
        if valor in inverso:
            # Lo convertimos a lista si es la primera vez que se repite
            if not isinstance(inverso[valor], list):
                inverso[valor] = [inverso[valor]]
            inverso[valor].append(clave)

            # Registramos duplicados para el aviso
            duplicados.setdefault(valor, []).append(clave)
        else:
            # Caso sin duplicados
            inverso[valor] = clave

    # Mostramos el aviso si hay duplicados
    if duplicados:
        print("Aviso: hay claves repetidas al invertir el diccionario.\n")
        for valor in duplicados:
            claves = inverso[valor]
            print(f"El valor '{valor}' está asociado a las claves: {claves}")

    return inverso
```

```python
dic = {"a": 1, "b": 2, "c": 1, "d": 2}
resultado = diccionario_inverso(dic)
print("\nResultado:", resultado)
```

1. Comptador i endreçador de paraules d'un text.

El client va quedar content amb el comptador de paraules, però ara vol llegir arxius TXT i que calculi la freqüència de cada paraula ordenades dins de les entrades habituals del diccionari segons la lletra amb la qual comencen, és a dir, les claus han d'anar de la A a la Z i dins de la A hem d'anar de la A la Z.

```python
def contar_palabras_por_letra(nombre_archivo):
    """
    Lee un archivo de texto y devuelve un diccionario con la frecuencia de palabras
    agrupadas por su letra inicial (A-Z).
    """

    # Abrimos y leemos el archivo y lo convertimos todo a minúsculas
    with open(nombre_archivo, 'r', encoding='utf-8') as f:
        texto = f.read().lower()

    # Extraemos palabras, incluyendo tildes, ñ y diéresis
    import re
    palabras = re.findall(r"[A-Za-zÁÉÍÓÚáéíóúñÑüÜ]+", texto)

    # Contamos cuántas veces aparece cada palabra
    contador = {}
    for palabra in palabras:
        contador[palabra] = contador.get(palabra, 0) + 1

    # Creamos la estructura A-Z y agrupamos por la letra inicial
    dicc_a_z = {letra: {} for letra in "ABCDEFGHIJKLMNOPQRSTUVWXYZ"}

    for palabra, frecuencia in contador.items():
        inicial = palabra[0].upper()

        if inicial in dicc_a_z:
            dicc_a_z[inicial][palabra] = frecuencia

    # Ordenamos alfabéticamente cada subdiccionario
    dicc_a_z = {
        letra: dict(sorted(subdicc.items()))
        for letra, subdicc in dicc_a_z.items()
    }

    return dicc_a_z
```

2. Conversió de tipus de dades.

El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir en flotants i l'altra on estan els elements que no es van poder convertir.

```python
lista = ['1.3', 'one', '1e10', 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

flotantes = []

def extraer_flotantes(obj):
    """
    Extrae valores numéricos convertibles a float desde estructuras anidadas,
    separando los convertibles y los que no en listas.
    """
    
    if isinstance(obj, (float, int)):
        flotantes.append(float(obj))
    elif isinstance(obj, str):
        try:
            f = float(obj)
            flotantes.append(f)
        except ValueError:
            pass
    elif isinstance(obj, (list, tuple)):
        for item in obj:
            extraer_flotantes(item)

extraer_flotantes(lista)

print(flotantes)
```

1. Generador de Contrasenyes

Explora el funcionament del mòdul random de la llibreria numpy.

En aquest punt, el client ha detectat un problema amb les contrasenyes que utilitzen els seus treballadors. Asdf1234, dates d’aniversari o similars. Per a solucionar-ho ens ha encarregat una funció de Python que generi contrasenyes més segures. La funció ha de dependre dels següents paràmetres:

longitud (int): Longitud de la contrasenya
majuscules (bool = True): Si hi ha d’aparèixer majúscules
minuscules (bool = True): Si hi ha d’aparèixer minúscules
numeros (bool = True): Si hi ha d’aparèixer números
signes (bool = False): Si hi ha d’aparèixer caràcters especials (,-$? o similars)
Així doncs, si executem la funció de la següent manera:

crear_contrasenya(10, True, True, True, True)

Hauríem d’obtenir un output (que hem de poder guardar) de l’estil:

9Er,5Vn8P$

Assegura’t que es compleixin tots els criteris, i que aquestes contrasenyes siguin realment aleatòries.

(EXTRA) Explora com podríem fer que la funció copiés la contrasenya automàticament al porta papers de l’ordinador (com si l’haguéssim seleccionat i fet ctrl+copy).

```python
import numpy as np

MAYUSCULAS = np.array([chr(i) for i in range(ord('A'), ord('Z') + 1)])
MINUSCULAS = np.array([chr(i) for i in range(ord('a'), ord('z') + 1)])
NUMEROS = np.array([str(i) for i in range(10)])
SIMBOLOS = np.array(list("!@#$%^&*()-_=+,.?/"))

def crear_contrasenya(longitud, mayusculas=True, minusculas=True, numeros=True, simbolos=False):
    """
    Genera una contraseña aleatoria con NumPy según longitud y tipos de caracteres.
    """
    grupos = []

    if mayusculas:
        grupos.append(MAYUSCULAS)
    if minusculas:
        grupos.append(MINUSCULAS)
    if numeros:
        grupos.append(NUMEROS)
    if simbolos: 
        grupos.append(SIMBOLOS)
    
    if not grupos:
        raise ValueError("Debe activarse al menos un tipo de carácter.")
    
    if longitud < len(grupos):
        raise ValueError("La longitud es demasiado pequeña para los tipos seleccionados.")

    contrasenya = []

    for grupo in grupos:
        contrasenya.append(np.random.choice(grupo))

    pool = np.concatenate(grupos)

    while len(contrasenya) < longitud:
        contrasenya.append(np.random.choice(pool))

    contrasenya = np.array(contrasenya)
    np.random.shuffle(contrasenya)

    return "".join(contrasenya)

print(crear_contrasenya(10, True, True, True, True))
```
